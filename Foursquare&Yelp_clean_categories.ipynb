```python
import pandas as pd
import numpy as np
```

```python
#Open Foursquare and Yelp DataFrames
foursquare_venues_new=pd.read_csv('foursquare_venues_new.csv',index_col=[0])
yelp_venues=pd.read_csv('yelp_venues.csv',index_col=[0])
```

```python
#Take care of duplicating nbhoods
yelp_venues_new=yelp_venues.replace('Chelsea', 'Chelsea M')
```

```python
#Create the combined DataFrame for venues
venues=pd.concat([foursquare_venues_new, yelp_venues_new], axis=0, ignore_index=True)
```

```python
venues
```

```python
#Number of venues that appear several times
venues.duplicated(subset=['Venue Latitude', 'Venue Longitude']).sum()
```

```python
#Duplicates that appear in same neighborhood several times (category may vary) -need to be removed
venues.duplicated(subset=['Neighborhood','Venue Latitude','Venue Longitude']).sum() 
```
