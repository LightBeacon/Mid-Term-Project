```python
import pandas as pd
import numpy as np
```

After we scaped the events data from yelp API. we divided them into two types:
    1- Future events 
    2- Past events

```python
future_events=pd.read_csv('yelp_future_events.csv')
```

```python
future_events['Start Date'].max()
```

```python
past_events=pd.read_csv('yelp_past_events.csv')
```

```python
past_events.head()
```

```python
events=pd.concat([past_events, future_events],ignore_index=True)
```

```python
# Group by the dataframe according to the neighborhoods
events=events.groupby(['Neighborhood']).count()
```

```python
events
```

```python
events.loc[events.duplicated(subset=['Event','Event Latitude', 'Event Longitude', 'Start Date'])==True]
```

```python
number_of_events=events.groupby(['Neighborhood']).count().sort_values(by=['Event'], ascending=False)
```

```python
number_of_events['Number of Events']=number_of_events['Event']
```

```python
# drop columns not in concern
number_of_events=number_of_events.drop(['Unnamed: 0', 'Event Latitude', 'Event','Event Longitude', 'Event Category', 'Start Date', 'End Date'], axis=1)
```

```python
number_of_events
```

```python
# Exporting the Yelp events data into csv file

working_dataframe=pd.read_csv('working_dataframe_7.csv')
```

```python
working_dataframe=working_dataframe.merge(number_of_events, on='Neighborhood', how='left')
```

```python
working_dataframe
```

```python
working_dataframe.loc[working_dataframe['Number of Events'].isna()].count() #222/306 (72%) nbhoods do not have events.
```

```python
# Replace NaN with value of 0
working_dataframe['Number of Events']=working_dataframe['Number of Events'].replace(np.nan, 0)
```

```python
working_dataframe.to_csv('working_dataframe_2.csv', index=False)
```
