### Function to pull the data from Uber dataset for each neighborhoods of NYC
Be careful this function will take long time, so if you want test it, take a small sample of uber rides, like: 1000 only

```python
#Assign neighborhood to each ride
uber_sep14['LL'] = list(zip(uber_sep14.Lat, uber_sep14.Lon))
raduis=908
working_dataframe['LL']=list(zip(working_dataframe.Latitude,working_dataframe.Longitude))
uber_sep14['Neighborhood']=0
for n in tqdm_notebook(range(len(uber_sep14['LL']))):
               point=uber_sep14['LL'][n]
               for m in range(len(working_dataframe['LL'])):
                              center=working_dataframe['LL'][m]
                              distance=geopy_distance.distance(point, center).meters  
                              if distance<raduis:                            
                                             uber_sep14['Neighborhood'][n]=working_dataframe['Neighborhood'][m+1]
            
```

```python
# Join uber data with the working dataframe
Uber_rides=pd.read_csv('Uber_rides.csv')
working_dataframe=pd.read_csv('working_dataframe_9.csv')
```

```python
df_uber_rides = Uber_rides.drop(columns=['Date/Time','Lat','Lon','Base'])
df_uber_rides = df_uber_rides.rename(columns={'LL': 'Daily_uber_rides'})
df_uber_rides=df_uber_rides.iloc[1: , :]
df_uber_rides
```

```python
df_al = pd.merge(working_dataframe,df_uber_rides,on='Neighborhood',how='outer')
df_al
```

```python
df_al['Daily_uber_rides']=df_al['Daily_uber_rides'].fillna(0)
```

```python
df_al=df_al.drop(columns=['Number_of_Rides'])
df_al
```

```python
df_al.to_csv('final_dataframe.csv', index=False)
```
